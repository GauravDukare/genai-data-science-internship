```python
def defangIPaddr(address):
        res = ""
        for i in address:
            if i == ".":
                res += "[.]"
            else:
                res += i

        return res

address1 = "1.1.1.1" 
address2 = "255.100.50.0"

print(defangIPaddr(address1))
print(defangIPaddr(address2))
```

```python
def findNumbers(nums):
        res = 0
        for i in nums :

            count = 0
            while i != 0:
                i = i // 10
                count += 1

            if count % 2 == 0:
                res += 1

        return res

nums1 = [12,345,2,6,7896]
nums2 = [555,901,482,1771]

print(findNumbers(nums1))
print(findNumbers(nums2))
```

```python
def numIdenticalPairs(nums):
        res = 0
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if i < j and nums[i] == nums[j]:
                    res += 1
             
        return res

nums1 = [1,2,3,1,1,3]
nums2 = [1,1,1,1]
nums3 = [1,2,3]

print(numIdenticalPairs(nums1))
print(numIdenticalPairs(nums2))
print(numIdenticalPairs(nums3))
```

```python
def smallerNumbersThanCurrent(nums):
        res = []
        for i in range(len(nums)):
            count = 0
            for j in range(len(nums)):
                if i != j and nums[j] < nums[i]:
                    count += 1
            res.append(count)

        return res

nums1 = [8,1,2,2,3]
nums2 = [6,5,4,8]
nums3 = [7,7,7,7]

print(smallerNumbersThanCurrent(nums1))
print(smallerNumbersThanCurrent(nums2))
print(smallerNumbersThanCurrent(nums3))
```

```python
def subtractProductAndSum(n):
        prod = 1
        add = 0
        num = n

        while num != 0:
            r = num % 10
            add += r
            prod *= r
            num //= 10

        return prod - add

n1 = 234
n2 = 4421

print(subtractProductAndSum(n1))
print(subtractProductAndSum(n2))
```

```python
def xorOperation(n: int, start: int) -> int:
        res = 0
        for i in range(n):
            res ^= start + 2 * i

        return res

n1 = 5 
start1 = 0

n2 = 4 
start2 = 3

print(xorOperation(n1, start1))
print(xorOperation(n2, start2))
```
