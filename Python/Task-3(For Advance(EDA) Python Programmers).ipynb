```python
def maxProduct(nums: int) -> int:
        res = 0

        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if ((nums[i]-1) * (nums[j]-1)) > res:
                    res = (nums[i]-1) * (nums[j]-1)

        return res

nums1 = [3,4,5,2]
nums2 = [1,5,4,5]
nums3 = [3,7]

print(maxProduct(nums1))
print(maxProduct(nums2))
print(maxProduct(nums3))
```

```python
def numTeams(rating):
        n = len(rating)
        count = 0

        for j in range(n):
            leftl = leftg = rightl = rightg = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    leftl += 1
                elif rating[i] > rating[j]:
                    leftg += 1

            for k in range(j + 1, n):
                if rating[k] < rating[j]: 
                    rightl += 1
                elif rating[k] > rating[j]:
                    rightg += 1

            count += leftl * rightg + leftg * rightl

        return count


rating1 = [2,5,3,4,1]
rating2 = [2,1,3]
rating3 = [1,2,3,4]

print(numTeams(rating1))
print(numTeams(rating2))
print(numTeams(rating3))
```

```python
def busyStudent(startTime, endTime, queryTime):
        res = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                res += 1

        return res

startTime1 = [1,2,3]
endTime1 = [3,2,7]
queryTime1 = 4

startTime2 = [4]
endTime2 = [4]
queryTime2 = 4

print(busyStudent(startTime1, endTime1, queryTime1))
print(busyStudent(startTime2, endTime2, queryTime2))
```

```python
def numberOfSteps(num) :
        res = 0

        while num != 0:
            if num % 2 == 0:
                num = num / 2
            else:
                num = num - 1
            res += 1

        return res

num1 = 14
num2 = 8
num3 = 123

print(numberOfSteps(num1))
print(numberOfSteps(num2))
print(numberOfSteps(num3))
```

```python
def countBits(n):
        dp = [0]  * (n + 1)
        offset = 1

        for i in range(1, n+1):
            if offset * 2 == i:
                offset = i

            dp[i] = 1 + dp[i - offset]

        return dp

n1 = 2
n2 = 5

print(countBits(n1))
print(countBits(n2))
```
