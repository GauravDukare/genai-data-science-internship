```python
def runningSum(nums):
        running_sum = []
        temp = 0
        for i in nums:
            temp += i
            running_sum.append(temp)

        return running_sum

nums1 = [1,2,3,4]
nums2 = [1,1,1,1,1]
nums3 = [3,1,2,10,1]


print(runningSum(nums1))
print(runningSum(nums2))
print(runningSum(nums3))
```

```python
def shuffle(nums, n):
    res = []

    for i in range(n):
        res.append(nums[i])
        res.append(nums[i+n])

    return res

nums1 = [2,5,1,3,4,7]
n1 = 3

nums2 = [1,2,3,4,4,3,2,1]
n2 = 4

nums3 = [1,1,2,2]
n3 = 2


print(shuffle(nums1, n1))
print(shuffle(nums2, n2))
print(shuffle(nums3, n3))
```

```python
def kidsWithCandies(candies, extraCandies):
        greatest = max(candies)
        res = []

        for i in candies:
            if i + extraCandies >= greatest:
                res.append(True)
            else:
                res.append(False)

        return res


candies1 = [2,3,5,1,3]
extraCandies1 = 3

candies2 = [4,2,1,1,2]
extraCandies2 = 1

candies3 = [12,1,12]
extraCandies3 = 10

print(kidsWithCandies(candies1, extraCandies1))
print(kidsWithCandies(candies2, extraCandies2))
print(kidsWithCandies(candies3, extraCandies3))
```
